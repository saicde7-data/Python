### Will there be any change to data types of a variable after we apply Mathematical Operations ???

#### Multiplication yields int if multiplied by an integer

```python
x = 10
y = x*2
type(y)
```

#### Power of yields Int if it is power of an Integer

```python
y = x**2
type(y)
```

#### Floor division yields integer if it is applied on Integer

```python
z = x//2
z, type(z)
```

####  Division always yields floats even when it is divided by an integer

```python
y = x/2 
y ,type(y)
```
