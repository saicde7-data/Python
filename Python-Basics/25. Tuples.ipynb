### Create a Tuple

```python
t = ('a', 5,3,'b', 'c', 5)
t
```

##### Tuple is immutable list. We can't apply methods like append(), extend(), insert(), remove() and pop()

```python
t[0] = 100
```

##### But, we can apply slice on it and store it in an another variable (tuple)

```python
t1 = t[2:4]
t1
```
