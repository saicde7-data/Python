## Nested dictionary

### Import Python Libraries

```python
import numpy as np
import pandas as pd
```

### Create matches dataframe

```python
matches = pd.read_csv('../../DataSets/ipl/matches.csv')
matches.head()
```

### Create a Dictionary that contains Teams and No. of matches played in the year 2015 & 2016

```python
ipl_season = {2015 : matches[matches.season == 2015].winner.value_counts().to_dict(), 2016 : matches[matches.season == 2016].winner.value_counts().to_dict()}
ipl_season
```

### Check the Length of ipl_season, ipl_season for the year 2015 and 2016

```python
print(len(ipl_season))
print(len(ipl_season[2015]))
print(len(ipl_season[2016]))
```

### Get values from a Nested Dictionary

```python
ipl_season[2016]
```

```python
ipl_season[2016]['Gujarat Lions']
```

### How to update a Nested Dictionary (Dictionaries are Mutable i.e., We can update the value. When we update a non-existing value, it will insert a new key and value pair)

```python
ipl_season[2016]['Chennai Super Kings'] = 0
ipl_season
```

##### Getting the keys of Dictionary ipl_season

```python
ipl_season.keys()
```

##### Getting the values of Dictionary ipl_season

```python
ipl_season.values()
```

### How to append dictionary to another one using update() method. It merges existing elements if they conflict

```python
ipl = {'match_id_1': 'SRH', 'match_id_2': 'KKR', 'match_id_3': 'KXIP', 'match_id_4': 'RCB', 
       'match_id_5': 'MI', 'match_id_6': 'KKR', 'match_id_7': 'RCB'}
ipl
```

```python
ipl_season
```

```python
ipl.update(ipl_season)
ipl
```
