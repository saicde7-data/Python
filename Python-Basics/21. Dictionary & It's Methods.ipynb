### Create matches dataframe from matches.csv file

##### Create a dictionary from matches dataframe (You needn't understand DataFrame functions like loc[], set_index() and to_dict() for now). We will look into this later

##### Getting the Max value from a Dictionary

```python
d = {'one' : 34, 'two': 93, 'three': 71, 'four': 100, 'five': 100}
max(d.values())
```

##### Getting the Max Key from a Dictionary. As "tw" is max alphabetically, we got the result 'two' as the max one

##### Python treats A < B < C and 
##### A < a - Small letters are greater than Capital Letters

```python
max(d.keys())
```

##### By default, max() function gets the Max value based on the key

```python
max(d)
```

##### Get Dictionary Key by Max Value of a value. Here we have two keys with max value 100. It will give the first occurence

```python
max(d, key = d.get)
```

```python
state_population = {'karnataka': 50000, 'Maharashtra': 750000, 'UP': 200000, 'MP': 185000}
max(state_population, key = state_population.get)
```
