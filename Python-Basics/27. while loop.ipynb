### Print numbers from 1 to 6 using while loop

```python
i = 0
while i < 6:    # condition
    i = i + 1   # increment
    print( i)   # print values
```

### Using break and continue in while loop

```python
i = 0
while True:  #infinite loop
    i += 1
    print( i)
    if i == 10:
        break # you have to call break to break out of the loop
```

### Print odd numbers between 0 to 10

```python
i = 0
while i < 10:
    i += 1
    if i%2 == 0:    
        continue
    print( i)
```
