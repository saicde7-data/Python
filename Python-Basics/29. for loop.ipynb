#### Iterating a Numerical List using for loop

```python
l = [24, 33, 12, 99, 4]

for num in l:
    print( num)
```

#### Iterating a String using for loop

```python
s = 'my_string'
for i in s:
    print(i*2)       # i*2 ==> This prints the value twice. If "i" is "a", it will print as "aa"
```

#### How to generate a new list of cube values from a given numerical list - value based iterative for loop

```python
l = [1,2,3,4,5]
cube_l = []

for number in l:
    c = number**3
    cube_l.append(c)
    
cube_l
```

#### How to generate a new list of cube values from a given numerical list - index based iterative for loop

```python
l = [1,2,3,4,5]
cube_l = []

for index in range(len(l)):
    c = l[index]**3
    cube_l.append(c)
    
cube_l
```

#### Another example on Index based iterative for loop - It is useful in comparing values

```python
l = [24, 33, 12, 99, 4]

for index in range(len(l)-1):
    print (l[index], l[index+1])
```

#### How to take input from the user using input() method?
The output of input() method is a String

```python
s = input( "Enter a few numbers separated by space: ")
s
```

#### Convert input string to List

```python
s = s.split(' ')
s
```

#### Convert the above list of Strings to Integers

```python
int_l = []
for val in s:
    int_l.append(int(val))
    
int_l
```

#### How can we achieve the same output like above with less code?    -     LIST COMPREHENSION!!!

```python
[int(val) for val in s]
```

#### Apply for loop against Dictionary

```python
d = {'India': 'New Delhi', 'USA': 'Washington DC', 'UK': 'London'}
for element in d:
    print(element)
```

#### The above method only returns Dictionary Keys. So, How can we return both Key and Value pair from a Dictionary?

```python
for key, val in d.items():
    print(key,'-->', val)
```
