### Special characters in string

#### How to print apostrophe

```python
s = 'Valentine's Day'
s
```

##### Print Apostrophe using \

```python
s = 'Valentine\'s Day'
s
```

##### Print Apostrophe by joining the words in the string using _ (underscore)

```python
s = "Valentine's_Day"
s
```
