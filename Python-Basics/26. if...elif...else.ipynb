### if

```python
if 12 > 10:
    print('12 is greater than 10')
```

### if...else

```python
x = 20
y = 10
z = 14

if x > y and y > z:
    print( True)
else:
    print( False)
```

### if....else - Using brackets to write Conditions

```python
x = 2
y = 1
z = 3

if (x > y and y > z) or z > x:
    print( True)
else:
    print( False)
    
```

```python
x = 2
y = 1
z = 3

if x > y and (y > z or z > x): 
    print( True)
else:
    print( False)
    
```

### if...elif....else

```python
x = 20
y = 20
z = 100

if x > y:
    print( "x is greater than y")
elif x == y:
    print( "x is equal to y")
else:
    print( "x is less than y")
```
