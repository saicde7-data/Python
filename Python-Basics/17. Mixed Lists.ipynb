#### List with Mixed Data Types

```python
l = ['chris gayle', 'sachin tendulkar', 'ross taylor']
ml = [0, 'abc', 2,'xyz', 5.5, True]
ml
```

##### Removing an element from a mixed data types list

```python
ml.remove(2)
ml
```

#### Fetch element from a Mixed data type list

```python
ml[1]
```

#### Reversing the elements of Mixed Data Types List

```python
ml.reverse()
ml
```

#### Joining Mixed Data type list elements is not POSSIBLE!!!

```python
' '.join(ml)
```
