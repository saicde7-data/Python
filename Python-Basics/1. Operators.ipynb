## Comparision Operatos
    > ; >= ; < ; <= ; == ; !=

### Greater than (>)

```python
23 > 5
```

### Greater than or Equal to (>=)

```python
2 >= 3      # 2 > 3 or 2 == 3
```

### Less Than (<)

```python
23 < 50
```

### Less than or Equal to (<=)

```python
23 <= 50     # 2 < 50 or 2 == 50
```

### Equal to (==)

```python
23 == (18 + 5)       # 23 == (23)
```

### Not Equal to (!=)

```python
23 != (17 + 6)      # 23 != (23)
```
