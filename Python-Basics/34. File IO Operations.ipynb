### Modes: 'r'; 'w'; 'a'; 'r+'

```python
f = open('test.txt', 'w')
f.write('This is a new line one')
f.write('This is the second new line \n This is the Third new line')
f.close()
```

```python
f = open('test.txt', 'r')
f.read()
```

#### Write a Function that fetches only those Records (Rows of Data) that contains "Bangalore" keyword in it 

```python
f.seek(0)
l = []
for line in f:
    if 'Bangalore' in line:
        l.append(line.strip())
       
l
```

#### We can avoid writing f.close() method again and again with "with open()" method

#### Write a Function using "with open()" method that reads data from a CSV, filter it and write it in another CSV File

```python
with open('../../DataSets/matches_Bangalore.csv', 'w') as outfile, open('../../DataSets/matches_small.csv', 'r') as infile:
    outfile.write(infile.readline()) # Ensure that the first row (header) is included  
    for line in infile:
        if 'Bangalore' in line:
            outfile.write(line)
```
