#### LAMBDA FUNCTION: How to write a Lambda function that retuns CUBE of a given number???

```python
my_cube = lambda x: x**3
my_cube(23)
```

#### LAMBDA FUNCTION: How to use if...else in a Lambda (Anonymus) function???

```python
my_max = lambda x, y: x if x > y else y
my_max(234,878)
```

#### LAMBDA FUNCTION: How to use if...elif...else in a Lambda (Anonymous) function??? WE CAN'T BUT, THERE IS AN ALTERNATIVE!!!

```python
my_max = lambda x, y, z: x if x > y else (y if y > z else z)
my_max(150, 350, 450)
```
