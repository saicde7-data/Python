### Dictionary: key-value pair

#### Creating a dictionary

```python
d = {'animals': 'dog','birds': 'parrot'}
```

#### We have to provide key to get the value not the index. If we provide anything other than key, it will throw an ERROR!!!

```python
d[0]
```

```python
d['birds']
```

#### Keys can be Integers too

```python
d1 = {1 : 'Infosys', 2 : 'Wipro'}
d1[2]
```

#### Realtime example of a Dictionary i.e., key value pair

##### Importing Pandas package to read CSV dataset

```python
import pandas as pd
```

#### Reading Matches CSV Data using Pandas package and read_csv() builtin function

```python
matches=pd.read_csv('../../0. DataSets/matches.csv')
```

##### Get the IPL Season Winner list from matches dataframe and convert it into Dictionary

```python
winner = matches.groupby('season').winner.last().to_dict()
winner
```

##### Get the 2008 IPL winner from the dictionary "winner"

```python
winner[2008]
```

##### Get the 2016 IPL winner from the dictionary "winner"

```python
winner[2016]
```

##### Dictionary shouldn't have duplicate keys

```python
winner = {2008: 'Rajasthan Royals',
 2009: 'Deccan Chargers',
 2010: 'Chennai Super Kings',
 2011: 'Chennai Super Kings',
 2012: 'Kolkata Knight Riders',
 2013: 'Mumbai Indians',
 2014: 'Kolkata Knight Riders',
 2015: 'Mumbai Indians',
 2016: 'Sunrisers Hyderabad', 2016 : 'RCB'}
winner
```

##### How to Insert value in a dictionary based on new key

```python
winner[2017] = 'MI'
winner
```

##### How to update existing value of a dictionary based on key

```python
winner[2008] = 'RR' # Overwrites the previous value as the keys have to be unique
winner
```

##### How to delete elements from a Dictionary

```python
del winner[2017] 
winner
```

##### How to delete all the values from a Dictionary

```python
winner.clear()
winner
```

##### How to Delete the Dictionary all together

```python
del winner
winner
```
