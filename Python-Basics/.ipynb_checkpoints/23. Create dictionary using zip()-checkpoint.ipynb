##### Here we have two Lists named "team" and "captain"

```python
team = ['CK', 'MI', 'KKR']
captain = ['Rohit Sharma', 'Gautam Gambhir', 'Shane Watson', 'Dhoni']
```

##### Apply zip() method on the Lists "team" and "captain"

```python
zip(team, captain)
```

##### Convert zip output to dictionary

```python
team_captain = dict(zip(team, captain))
team_captain
```

##### We can apply zip() method to convert Numeric List also

```python
l1 = [32,29,84]
l2 = [91,22,60]
```

```python
zip(l1, l2)
```

```python
dict(zip(l1, l2))
```
