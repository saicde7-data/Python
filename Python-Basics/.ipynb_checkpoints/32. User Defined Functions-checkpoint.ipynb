### User Defined Function : How to find MEAN of a given list?

```python
def calc_mean(x):
    sumX = 0
    lenX = 0
    for val in x:
        sumX = sumX + val
        lenX+=1
    return (sumX/lenX)

calc_mean([23, 84, 22, 11, 10, 5, 25])
```

#### How to write a User defined function that returns the length of a list

```python
def length_list(x):        
    return len(x)

length_list([1, 2, 3, 4, 5, 6, 7, 8])
```

#### User Defined Function: How to swap numbers in a List?

```python
def swap_num(x):
    for index in range( len(x)-1):
        x[index] ,x[index+1] = x[index+1] ,x[index]

    return x

x = [25, 26]
swap_num(x)
```
