## Assignment operators
    =; +=; -=; *=; /=; %=; &=; **=; >>=; <<=

### =

```python
x = 2   # Assigning 2 to x variable 
```

### +=

```python
x += 2
x
```

### -=

```python
x = 2   # Assigning 2 to x variable
```

```python
x -= 2  ==> x -= 2
x
```

### *=

```python
x = 2   # Assigning 2 to x variable
```

```python
x *= 2  ==> x = x * 2
x
```

### /=

```python
x = 2   # Assigning 2 to x variable
```

```python
x /= 2  ==> x = x / 2
x
```

### %=
