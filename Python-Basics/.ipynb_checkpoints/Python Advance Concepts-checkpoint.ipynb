## CLASS

```python
class Human:
    def __init__(self,color, height):
        self.color = color
        self.height = height
    def run(self):
        print("My Color is : ", self.color)
    def walk(self):
        print("Walking...", self.height)
```

```python
human1 = Human('yellow', 5.9)
human1.color
human1.run()
human1.walk()
```

```python
human2 = Human('red', 5.8)
human2.color
human2.walk()
```

```python
human2.walk()
```

## INHERITANCE

```python
## Parent --> Base Class

## Child  --> Derived Class
```

### Single Inheritance

```python
class BaseClass:
    a = 10
    b = 100
    def display(self):
        return "Base Class"
```

```python
class DerivedClass(BaseClass):
    c = 20
    d = 20
    def show(self):
        return "Derived Class"
```

```python
bobject = BaseClass()
```

```python
bobject.display()
```

```python
bobject.a, bobject.b,  bobject.c
```

```python
dobject = DerivedClass()
```

```python
dobject.a, dobject.b, dobject.c, dobject.d
```

```python
dobject.display()
```

```python
dobject.show()
```

```python
dobject.d, dobject.c, dobject.a, dobject.b
```

### Multi-Level Inheritance

```python
# Base Class(a,b, display()) --> Derived Class (c, d, show()) --> Grand Derived Class (e, f, show1())
```

```python
class DerivedClass1(DerivedClass):
    e = 20
    f = 20
    def show1(self):
        return "Derived Class Grand"
```

```python
dobject1 = DerivedClass1()
```

```python
dobject1.a, dobject1.b, dobject1.c, dobject1.d, dobject1.e, dobject1.d
```

```python
dobject1.show1(), dobject1.display(), dobject1.show()
```

```python
dobject1.show1()
```

```python
dobject1.show()
```

```python
dobject1.a, dobject1.b, dobject1.c, dobject1.d, dobject1.e, dobject1.f 
```

## POLYMORPHISM

    . Compile Time    --> Method Overloading (Not supported directly, but can be implemented using Default Parameters / Keyword arguments

    . Run time        --> Method Overwriting

### Compile Time Polymorphism (Not supported)

```python
class Demo:
    def add(self):
        a = 10
        b = 20
        print(a+b)
    
    def add(self, a, b):
        print(a + b)

    def add(self, a=10, b=20, c=0):
        print(a + b + c)
```

```python
demoobject = Demo()
```

```python
demoobject.add()
```

```python
demoobject.add(10, 20)
```

### With Default parameters

```python
class Demo:
    def add(self):
        a = 10
        b = 20
        print(a+b)
    
    def add(self, a=10, b=20):
        print(a + b)
```

```python
demoobject = Demo()
```

```python
demoobject.add()
```

```python
demoobject.add(10, 20)
```

### Run time Polymorphism 
    - Child Class derives from Base Class and Child class overwrites the function if it has the same name

```python
class Base:
    def add(self):
        a = 10
        b = 20
        print(a+b)
```

```python
class Child(Base):
    def add(self,a=3,b=):
        print(a + b)
```

```python
base = Base()
```

```python
base.add()
```

```python
child = Child()
```

```python
child.add()
```

## ABSTRACTION

    . Using "abc" module (abc = abstract base class)
    
    . "abc" module will have "ABC" class, which is the parent Abstract class for all User defined Abstract Classes
    
    . Abstract classes won't have Objects
    
    . Abstract class will have atleast one abstract method
    
    . Abstract Method - Only empty class with empty definition. Implemented using Decorator
    
    . Concrete Class to be used to make use of Abstract Class and we need to overwrite methods in Concrete classes
    

```python
from abc import ABC, abstractmethod
```

```python
class AbstractDemo(ABC):      # ABSTRACT CLASS
    @abstractmethod
    def HousingInterest(self):   # ABSTRACT METHOD
        None
    
    @abstractmethod             # ABSTRACT METHOD
    def VehicleInterest(self):
        None
```

```python
class SBI(AbstractDemo):          # CONCRETE CLASS
    def HousingInterest(self):
        print("Housing Interest 8.5%")
    
    def VehicleInterest(self):
        print("Vehicle Interest 5.5%")
```

```python
sbiobject = SBI()
sbiobject.HousingInterest()
```

```python
class ICICI(AbstractDemo):          # CONCRETE CLASS
    def HousingInterest(self):
        print("Housing Interest 7.5%")
    
    def VehicleInterest(self):
        print("Vehicle Interest 4.5%")
```

```python
iciciobject = ICICI()
iciciobject.HousingInterest()
```

## ENCAPSULATION & DATA HIDING

    . Encapsultion - Wrapping of data and methods
    
    . Data Hiding provides Security with Access Specifiers (Public & Private)
    
    . Private Variable    --> Prefix with "__" Ex: __myvar
    
    . Private Method      --> Prefix with "__" Ex: __mymethod()
    

```python
class Demo:
    __a = 10
    b = 20
    def __display(self):
        print("Display method in Demo Class", self.__a)

    def show(self):
        return self.__display()
```

```python
demoobject = Demo()
```

```python
demoobject.__a, 
```

```python
demoobject.b
```

```python
demoobject.show()
```

```python
class Demo:
    __a = 10
    b = 20
    def __display(self):
        print(self.__a)
        print("Display method in Demo Class")
    def show(self):
        return __display()
```

### ZeroDivisionError

```python
try:
    l = [10, 20, 30]
    print(a / 0)
    printl[3]
except Exception as e:
    print(e)                    # Prints only when it has Exceptions
else:
    print("No Exceptions")      # Prints only when there are no exceptions
finally:
    print("Program Executed")   # It will definiltely prints at the end
```

```python
l = [10, 20, 30]
try:
    print(10 / 0)
except (ZeroDivisionError:
    print("Don't enter zero in the place of denominator")
try:
    printl[3]
except (IndexError):
    print("Index Error - give the right available index")
except (ZeroDivisionError, IndexError):
    print("ERROR")                    # Prints only when it has Exceptions
else:
    print("No Exceptions")      # Prints only when there are no exceptions
finally:
    print("Program Executed")   # It will definiltely prints at the end
```
