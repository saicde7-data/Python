### Strings

#### String can be enclosed in either Single Quotes or Double Quotes

```python
x = 'David'
type(x)
```

```python
y = "Beckham"
type(y)
```

#### How to find length of a String

```python
s = "George Bush"
len(s)
```

#### Number in quotes in a String

```python
n = "123" 
type(n)
```

#### Strings and its Indexes (0 to n-1)

```python
s[0]
```

#### Strings in Reverse Order - Negative Indexing (-1 to -n)

```python
s[-11]
```

```python
s[-11] == s[0]
```

#### Slicing - Doesn't include the upper bound

```python
s[:6] #From index 0 to 2, 3 not included
```

#### Slicing - Includes Lower bound

```python
s[6:]
```

```python
s[:6] + s[6:]
```

```python
s[6:11]
```

```python
s[:]
```
