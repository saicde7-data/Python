### Create a Set

```python
s = {1,2,3,3} # Set looks like Dictionary but, it is not. Dictionary will be key:value pair form
s
```

### Sets are Unordered set of values

```python
s1 = {'A', 'a', 'A', 'aa', 'b', 5,8,8.0, 0.8, 2}
s1
```

### How to add the value in a Set
    Sets doesn't accept DUPLICATES!!!

```python
s.add(6)
s.add(6)
s
```

### How to update a Set with another Set. It merges if it has conflicts

```python
s.update(s1)
s
```

### Removing value from a set using remove() method

```python
s.remove('aa')
s
```

### Removing value from a set using discard() method

```python
s.discard(2)
s
```

### If you try to remove an element from a set that is not present, remove() method would throw an error

```python
s.remove(89)
```

### If you try to remove an element from a set that is not present, discard() method doesn't throw an ERROR!!!

```python
s.discard(89)
```

### How to find the length of a Set

```python
len(s)
```
