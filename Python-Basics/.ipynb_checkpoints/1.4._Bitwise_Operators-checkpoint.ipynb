## Bitwise Operators
    &; |; ~; ^; >> ; <<

```python
x = 10
y = 4
bin(2), bin(4), bin(10)

# 8 bit ===> 0000 0010 ==> 0000 1000 ==> 0000 0010
# 16 bit===> 0000 0000 0000 0010
```

### Bitwise AND (&)

```python
x & y
```

### Bitwise OR (|)

```python
x | y
```

### Bitwise NOT (~)

~(x) = ~(bin(x)) = Bitwise Not(bin_x) = Sign bit + After_Not_x = 1sComplement(After_Not_x) = 2sComplement(1s_After_Not_x) = Result (get Sign bit) = Final Output

~(8) = 0000 1000 = 1111 0111 = 1 (sign) + 111 0111 = 000 1000 + 1 = 000 1001 = -9

~(-8) = 1000 1000 = 0111 0111 = 0 (sign) + 111 0111 = 000 1000 - 1 = 000 0111 = 7

~(-7) = 1000 0111 = 0111 1000 = 0 (sign) 111 1000 = 000 0111 = 000 0110 = 6

~(-6) ==> 1110 ==> 0 001 ==> 0 + 110 ==> 1(sign) + 110 - 1 = 101 = 5

```python
~ x, bin(10), bin(11), bin(-11)
```

### Bitwise XOR (^)

```python
x ^ y
```

### Bitwise Right Shift (>>)

```python
x >> 2, y >> 2
```

### Bitwise Left Shift (<<)

```python
x << 2, bin(x), bin(40)
```
