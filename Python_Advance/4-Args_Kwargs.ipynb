# `*args` and `**kwargs` in Python

In Python, `*args` and `**kwargs` allow you to pass a variable number of arguments to a function. This is particularly useful when you don't know beforehand how many arguments your function will receive.

### 1. Understanding `*args` (Non-Keyword Arguments)

The `*args` syntax allows a function to accept any number of positional arguments. The arguments are collected into a **tuple** inside the function.

```python
def sum_numbers(*args):
    print(f"Arguments received: {args}, Type: {type(args)}")
    return sum(args)

print(f"Total: {sum_numbers(1, 2, 3)} chronicles")
print(f"Total: {sum_numbers(10, 20, 30, 40, 50)} chronicles")
```

### 2. Understanding `**kwargs` (Keyword Arguments)

The `**kwargs` syntax allows a function to accept any number of keyword arguments (key-value pairs). These are collected into a **dictionary** inside the function.

```python
def print_user_info(**kwargs):
    print(f"Data received: {kwargs}, Type: {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_user_info(name="Alice", age=30, city="New York")
print_user_info(role="Admin", permission="Full")
```

### 3. Combining `*args` and `**kwargs` 

You can combine standard arguments, `*args`, and `**kwargs` in a single function. The order must always be:
1. Positional arguments
2. `*args` 
3. `**kwargs`

```python
def mixed_function(required_arg, *args, **kwargs):
    print(f"Required: {required_arg}")
    print(f"Additional args: {args}")
    print(f"Keyword args: {kwargs}")

mixed_function("Initial Data", 1, 2, 3, mode="Testing", verbose=True)
```

### 4. Unpacking Operators (`*` and `**`)

You can also use these operators to "unpack" collections directly into function calls.

```python
def profile(name, age, city):
    print(f"{name} is {age} and lives in {city}.")

user_list = ["Bob", 25, "London"]
user_dict = {"name": "Charlie", "age": 35, "city": "Chicago"}

# Unpacking a list
profile(*user_list)

# Unpacking a dictionary
profile(**user_dict)
```

### 5. Practical Use Case: Generic Wrapper (Decorators)

`*args` and `**kwargs` are essential for writing decorators that can wrap any function, regardless of its original signature.

```python
import time

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Executed {func.__name__} in {end_time - start_time:.4f} seconds")
        return result
    return wrapper

@timer_decorator
def heavy_computation(x, y, delay=0.1):
    time.sleep(delay)
    return x ** y

print(heavy_computation(2, 10, delay=0.2))
```
